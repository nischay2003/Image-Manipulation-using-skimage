```python
import numpy as np
from skimage import io
from  skimage.color import rgb2gray
from skimage.util import invert
from skimage.io import imsave,imread
import matplotlib.pyplot as plt
```

```python
class Image_Manipulation():
    def _init_(self,image=None ):
        self.image = image
    def load(self):
        self.image=io.imread('Harry.jpeg')#reading the image
        self.image[0][0]=[255,255,255]
        plt.axis("off")#removing the axis
        plt.imshow(self.image)
        print("Your image is loaded")
        plt.show()
        
    def edit(self):
        while True:
            try:
                o1=int(input("Enter 1 for Horizontal Flip\nEnter 2 for Vertical Flip\nEnter 3 for Invert Colors\nEnter 4 for Convert to gray scale\n"))
                print()
            except:
                print("Enter the Correct option\n")
                continue
            else:
                if o1==1:
                    horizontal=self.image[::-1,:]#image is horizontally flipped
                    plt.axis("off")
                    plt.imshow(horizontal)
                    print("image is horizontall flipped")
                    plt.show()
                    
                elif o1==2:
                    vertical=self.image[:, ::-1]#image is vertically flipped
                    plt.axis("off")
                    plt.imshow(vertical)
                    print("image is vertically flipped")
                    plt.show()
                    
                elif o1==3:
                    inve=invert(self.image)#invert image
                    plt.axis("off")
                    plt.imshow(inve)
                    print("Image is Inverted")
                    plt.show()
                        
                elif o1==4:
                    grayimage=rgb2gray(self.image)#converting image to rgb2gray
                    plt.axis("off")
                    plt.imshow(grayimage, cmap="gray")
                    print("Your Image is converted to Grayscale")
                    plt.show()
                    
                
                
                else:
                    exit(0)
                    break
 #driver code               
                    
    def save(self):
        io.imsave(fname="E:\Github\Image Manipulation\saved\save.jpeg", arr=self.image)
        print("Your Image is successfully saved")
    
obj=Image_Manipulation()
while True:
    try:
        op=int(input("Enter 1 to Loading Image\nEnter 2 to Editing Image\nEnter 3 to Save the Image\n"))
        print()
    except:
        print("Enter the Correct Option\n")
        continue
    else:
        if op==1:
            obj.load()
            
            
        elif op==2:
            obj.edit()
            
        elif op==3:
            obj.save()
        else:
            exit(0)
            break
```
